# Importing libraries

```python
import pandas as pd
import quantstats as qs
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

# Get THE Financial market data for 4 different Public companies

```python
Amazon = yf.download('amzn', start="2018-03-01", end="2023-03-01").round(2)
Alibaba = yf.download('baba', start="2018-03-01", end="2023-03-01").round(2)
JD = yf.download('JD', start="2018-03-01", end="2023-03-01").round(2)
Shopify = yf.download('SHOP', start="2018-03-01", end="2023-03-01").round(2)

Shopify.head()
```

# Add the daily return percentage to the closing price

```python
Amazon['Daily_Return'] = Amazon['Close'].pct_change()
Amazon.reset_index(inplace=True)

Alibaba['Daily_Return'] = Alibaba['Close'].pct_change()
Alibaba.reset_index(inplace=True)

JD['Daily_Return'] = JD['Close'].pct_change()
JD.reset_index(inplace=True)

Shopify['Daily_Return'] = Shopify['Close'].pct_change()
Shopify.reset_index(inplace=True)

Shopify.head()
```

# Plotting Daily Returns for each stock

```python
# Amazon

Amazon.plot(x='Date', y='Daily_Return', figsize=(15,5));
plt.title('Amazon Daily Returns for the past 5 years')
plt.xlabel('Date')
plt.ylabel('Daily Return')

plt.show()
```

```python
# Alibaba

Alibaba.plot(x='Date', y='Daily_Return', figsize=(15,5));
plt.title('Alibaba Daily Returns for the past 5 years')
plt.xlabel('Date')
plt.ylabel('Daily Return')

plt.show()
```

```python
# JD

JD.plot(x='Date', y='Daily_Return', figsize=(15,5));
plt.title('JD Daily Returns for the past 5 years')
plt.xlabel('Date')
plt.ylabel('Daily Return')

plt.show()
```

```python
# Shopify

Shopify.plot(x='Date', y='Daily_Return', figsize=(15,5));
plt.title('Shopify Daily Returns for the past 5 years')
plt.xlabel('Date')
plt.ylabel('Daily Return')

plt.show()
```

# Combine the Adj close price for all companies and show in one plot

```python
Amazon_close = yf.download('amzn', start="2018-03-01", end="2023-03-01").round(2)
Alibaba_close = yf.download('baba', start="2018-03-01", end="2023-03-01").round(2)
JD_close = yf.download('JD', start="2018-03-01", end="2023-03-01").round(2)
Shopify_close = yf.download('SHOP', start="2018-03-01", end="2023-03-01").round(2)
```

```python
Amazon_close = Amazon_close['Adj Close']
Alibaba_close = Alibaba_close['Adj Close']
JD_close = JD_close['Adj Close']
Shopify_close = Shopify_close['Adj Close']
```

```python
adj_close = pd.concat([Amazon_close, Alibaba_close, JD_close, Shopify_close], join = 'outer', axis = 1)
adj_close.columns = ['Amazon', 'Alibaba', 'JD', 'Shopify']
adj_close
```

```python
adj_close.plot(figsize=(15,5));
plt.title('Daily Close Price for the past 5 years')
plt.xlabel('Date')
plt.ylabel('Close Price')

plt.show()
```
